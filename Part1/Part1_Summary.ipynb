## 셀 삽입
> - 바로 윗줄에 셀 생성 : `Ctrl + M A` (`A`: above)
> - 바로 아랫줄에 셀 생성 : `Ctrl + M B` (`B`: bottom)

## 셀 삭제
> - `Ctrl + M D` (`D`: delete)

## 셀 실행
> - 단일 셀 실행
  - 선택된 셀을 실행(커서는 그대로 남음) : `Ctrl + Enter`
  - 선택된 셀을 실행 후 다음 셀로 포커스 이동 : `Shift + Enter`
  - 선택된 셀 실행 후 다음줄에 새로운 셀 삽입 : `Alt + Enter`
> - 한번에 여러 셀 실행
  - `Ctrl/Shift` + `마우스 선택으로 실행하고자 하는 여러 셀 선택`
    - `Ctrl` : 원하는 셀을 하나씩 선택하여 여러 셀을 선택
    - `Shift` : 현재 선택된 셀부터 마우스로 클릭한 셀까지 한번에 선택
  - 우클릭 후 `선택된 여러 셀 실행`
  - 단축키 : `Ctrl + Shift + Enter`

# **5. 마크다운 문법**
[마크다운 공식 사이트](https://www.markdownguide.org/basic-syntax/)

## [헤더]<br>
> - `#` 으로 시작하는 텍스트이며, 제목을 말한다.
> - `#` 은 여섯개까지 가능하며, 갯수가 늘어날수록 헤더 크기가 작아진다.

```python
# 텍스트로 바꿔 확인해보자

# \# 하나인 헤더
## \# 두 개인 헤더
### \# 세 개인 헤더
#### \# 네 개인 헤더
##### \# 다섯 개인 헤더
###### \# 여섯 개인 헤더
```

## [수평선]
> - `-` or `*` or `_` 을 3개 이상 작성

---
******
_____

## [줄바꿈]
> - `<br>`을 입력하여 줄바꿈 (`br`: break)

Hello
Hello

Hello<br>Hello

## [강조하기]
> - 기울여 쓰기: `*` or `_`로 텍스트 감싸기
> - 두껍게 쓰기: `**` or `__`로 텍스트 감싸기
> - 취소선: `~~`로 텍스트 감싸기

*기울여 쓰기*<br>
**두껍게 쓰기**<br>
~~취소선~~<br>
*여러개를 **섞어서** ~쓰기도~ 가능하다*

## [인용구]
> - `>`으로 시작하는 텍스트
> - `>`의 갯수가 많아질 수록 하위수준으로 생성된다.

> 인용구의 생성
>> 하위 인용구의 생성
>>> 더 하위의 인용구 생성

## [리스트 목록]
> - 순서가 없는 리스트 : `-`, `+`, `*` 를 이용하여 생성
> - 순서가 있는 리스트 : `(숫자).` 형식을 이용하여 생성
> - 하위 수준의 리스트를 생성하려면 `tab`을 사용하여 들여쓰기를 한 후 생성

- 순서가 없는 리스트1
    - 순서가 없는 리스트2
        - 순서가 없는 리스트3

1. 순서가 있는 리스트1
    2. 순서가 있는 리스트2
        3. 순서가 있는 리스트3

## [특수 문자 표현]
> - 마크다운의 문자가 아닌 특수문자 그 자체를 사용하고 싶은 경우<br>백슬래시 `\`를 나타내고자 하는 특수문자 앞에 붙인다.

## 이미지 삽입
이미지를 삽입할 때는 html문법을 사용하는 것이 오류도 없고 수월하게 삽입할 수 있다.

아래의 코드는<br>'이미지가 OO에 있으니, 거기서 이미지 파일을 가져와서 코랩의 파일 셀 프레임에 띄어라'<br>라는 의미이다.

```python
%%html
<iframe src="https://img.mimint.co.kr/bbs/2020/05/27/C2005270925264610r.jpeg" 
width="800" height="1000" frameborder="0" allowfullscreen></iframe>
```

> src : source (소스 이미지의 주소)<br>
width : 가로 길이 (단위는 픽셀)<br>
height : 세로 높이

## 동영상 삽입
이미지 삽입과 같은 문법을 사용하지만, 소스부분(src)에 동영상 주소를 입력한다.

```python
%%html
<iframe src="https://youtu.be/Gh56nYRQ-FQ" 
width="560" height="315" frameborder="0" allowfullscreen></iframe>
```

단, 단순히 동영상 url을 복사하면 실행이 안 될 수도 있다.<br>
유튜브 같은 경우 개개인이 동영상을 공유하는 것을 막아두었기 때문이다.<br>

이때는 동영상 url을 복사하였을 때 마지막 슬레시(/)의 `동영상코드`를 복사한 뒤, 양식에 맞춰서 입력하면 실행이 된다.

```python
%%html
<iframe src="https://youtube.com/embed/Gh56nYRQ-FQ" 
width="560" height="315" frameborder="0" allowfullscreen></iframe>
```
