> - A `and` B : A 그리고 B (A와 B 둘 다 참)<br>
- A `or` B : A 또는 B (A와 B 중 하나는 참)<br>

```python
a = 1
b = 2

print((a>1) and (b>1))
print((a>1) or (b>1))
```

```python
not (a>1)
# False의 부정이므로 True이다.
```

> - `in`, `not in`

```python
b = 'Hello'
'e' not in b
```

#### if-else 문
> 조건이 True인 경우 if문, 그렇지 않은 경우 else문의 구문이 작동한다.<br>

**"*집이 5층보다 높으면 엘레베이터를, 그렇지 않으면 계단을 이용하라*"**

```python
floor = 10

# if 조건문에서 True를 만족하므로 if문 작동
if floor > 5 :
    print('엘레베이터를 탑니다.')

else :
    print('계단을 이용합니다.')
```

```python
floor = 2

# if 조건문에서 True를 만족하지 못하므로 else문 작동
if floor > 5 :
    print('엘레베이터를 탑니다.')

else :
    print('계단을 이용합니다.')
```

#### elif 문
> if문의 조건이 만족하면 if문,<br>
if문은 만족하지 않으나 elif문의 조건이 만족하면 elif문,<br>
모두 만족하지 않으면 else문의 구문이 작동한다.<br>

**"*배가 고프면 밥을 먹고, 배는 안 고프지만 목이 마르면 물을 마시고, 배도 안 고프고 목도 마르지 않으면 그냥 자라*"**

```python
status = ['thirst','full']

if 'hungry' in status :
    print('밥을 먹습니다.')

elif 'thirst' in status :
    print('물을 마십니다.')

else :
    print('잠을 잡니다.')
```

### 1-3-2. 반복문

#### while 반복문
> 조건을 만족하는 한 계속해서 구문을 반복<br>

**"*초콜릿을 5개 이상 먹었다면 그만 먹습니다.*"**

```python
hungry = True
eat_chocolate = 0

while hungry == True :
    eat_chocolate += 1
    print(f'초콜릿을 하나 먹습니다.(섭취한 초콜릿: {eat_chocolate}개)')

    if eat_chocolate >= 5 :
        hungry = False
        print('초콜릿을 너무 많이 먹었습니다.')
```

> `break`를 이용하여 중간에 멈출 수도 있다.

```python
hungry = True
eat_chocolate = 0
num_chocolate = 3

while hungry == True :
    eat_chocolate += 1
    num_chocolate -= 1
    print(f'초콜릿을 하나 먹습니다.(섭취한 초콜릿: {eat_chocolate}개)')

    if eat_chocolate >= 5 :
        hungry = False
        print('초콜릿을 너무 많이 먹었습니다.')

    if num_chocolate == 0 :
        print('초콜릿을 다 먹었습니다.')
        break
```

> 반복문을 끝낼 조건이 없다면 계속해서 구문이 돌아가는 **"*무한 루프*"**가 발생한다.

```python
eat_chocolate = 0

while True :
    eat_chocolate += 1
    print(f'초콜릿을 하나 먹습니다.(섭취한 초콜릿: {eat_chocolate}개)')
```

#### for 반복문
> 입력받은 값을 차례로 반복한다.

```python
food = ['Stake','Chicken','Noodle']

for eat in food :
    print(f'{eat} 을(를) 먹었습니다.')
```

> `continue`를 이용해 특정 조건을 스킵할 수 있다.

```python
food = ['Stake','Chicken','Noodle']

for eat in food :
    if eat == 'Noodle' :
        continue

    print(f'{eat} 을(를) 먹었습니다.')
```

> `break`를 이용해 while 반복문과 마찬가지로 중간에 반복을 멈출 수 있다.

```python
food = ['Stake','Chicken','Noodle']

for eat in food :
    if eat == 'Chicken' :
        break

    print(f'{eat} 을(를) 먹었습니다.')
```

> `range` 함수를 이용하여 반복 횟수를 설정할 수 있다.

### 1-3-3. 함수
> 같은 구문이 반복될 경우 구문을 함수로 만들어 사용할 수 있다.

```python
def sum(a,b) :
    return a + b

a = 1
b = 2
sum(a,b)
```

```python
# 함수를 사용하지 않을 때
food = ['Stake','Chicken','Noodle']

for eat in food :
    print(f'{eat} 을(를) 먹었습니다.')

food = ['Chocolate','Snack','Rice']

for eat in food :
    print(f'{eat} 을(를) 먹었습니다.')
```

```python
# 함수를 사용했을 때
def eating_food(food) :
    for eat in food :
        print(f'{eat} 을(를) 먹었습니다.')

# 함수 때문에 코드가 길어보일 수 있으나, 실제 코드는 아래 두 줄로 줄어드는 셈이다.
eating_food(['Stake','Chicken','Noodle'])
eating_food(['Chocolate','Snack','Rice'])
```

```python
# 함수의 리턴값으로 여러 개의 값을 받을 수도 있다.
def eating_food(food) :
    for eat in food :
        print(f'{eat} 을(를) 먹었습니다.')

    return len(food), food[-1]

num_food, last_food = eating_food(['Stake','Chicken','Noodle'])
print(f'총 {num_food} 개의 음식을 먹었습니다.')
print(f'마지막으로 {last_food} 을(를) 먹었습니다')
```

### 1-4-2. 파일 읽고 쓰기

#### 파일 생성하기
> - `open` : 새 파일을 만들 수 있다.
    - 파일이 한 번 오픈되면 반드시 `close`를 해주어야 한다.
    - close를 하지 않으면 파일이 저장되지 않고, 임시파일의 형태로 남게 된다.
- 파일 열기는 3가지의 모드가 있다.
    - `r` : 읽기 모드 (수정 불가)
    - `w` : 쓰기 모드
    - `a` : 추가 모드 (파일의 마지막에 내용 추가)

```python
f = open('test_file.txt', 'w')
f.close()
# 현재의 작업 경로에 새로운 'test_file.txt'가 생성되었음을 확인할 수 있다.
```

#### 파일 쓰기
> 위에서 언급한 `open`의 `w` 모드를 통해 파일의 내용을 변경할 수 있다.

```python
f = open('test_file.txt', 'w')
for i in range(1,11) :
    f.write(f'{i}번째 줄입니다.\n')

f.close()
# 위에서 만들었던 test_file.txt을 열어보면 내용이 입력되었음을 확인할 수 있다.
```

> 앞서 말했듯이 파일을 close 해주지 않으면 내용이 저장되지 않는다.

```python
f = open('test_file.txt', 'w')
for i in range(1,11) :
    f.write(f'{i}번째 줄입니다.')
# 파일에 들어가보면 내용이 입력되지 않았음을 확인할 수 있다.
```

#### 파일 읽기
> 위에서 언급한 `open`의 `r` 모드를 통해 파일의 내용을 확인할 수 있다.

```python
f = open('test_file.txt', 'r')
for line in f.readlines() :    # readlines() 함수는 내용을 한줄씩 읽는 함수이다.
    print(line)

f.close()
```

> - `with`문을 사용하면 파일을 좀 더 편리하게 다룰 수 있다.
    - with문은 파일을 연 뒤, with문 안의 구문이 모두 끝나면 파일을 자동으로 닫아준다.

```python
with open('test_file.txt', 'r') as f :
    for line in f.readlines() :
        print(line)
```

## 1-5. 그 밖의 고급 기능들

### 1-5-1. 리스트 내포 (List Comprehension)
> - '리스트 내포'는 리스트를 빠르고 쉽게 생성할 수 있는 파이썬의 기능 중 하나이다.
- 일반적으로 사용하는 for문 보다 간결하고 빠르게 리스트를 생성할 수 있다.


```python
# 0~9까지 담긴 리스트를 담아야 하는 경우

# 일반적인 for문을 사용한 경우
a = []
for i in range(10) :
    a.append(i)
print('a =',a)

# 리스트 내포를 사용한 경우
b = [x for x in range(10)]
print('b =', b)
```

> 뒤에 if 문을 넣어서 조건을 만들 수도 있다.

```python
# 짝수만 출력
odd_num = [x for x in range(10) if (x%2)==0]
odd_num
```

```python
# 성능 차이를 확인해보자.

# 'timeit'은 코랩에서 지원하는 모듈 중 하나이며, 코드를 여러 차례 실행한 후 실행시간을 출력해준다.
from timeit import timeit

def for_loop() :
    odd_list = []
    for n in range(100000) :
        if (n%2) == 0 :
            odd_list.append(n)

def list_comprehension() :
    odd_list = [x for x in range(100000) if (x%2)==0 ]
```

```python
# for문을 사용한 시간
print(timeit(for_loop, number=100))
print()

# 리스트 내포를 사용한 시간
print(timeit(list_comprehension, number=100))
```
