### 2-5-1. Pandas 자료구조 직접 만들기
> - Pandas에서는 `Series`와 `DataFrame` 두 가지의 자료구조를 사용한다.
    - Series : 일련의 객체를 담을 수 있는 1차원 배열 같은 자료구조
    - DataFrame : 표 같은 스프레드시트 형식의 자료구조
    - Series가 모여서 DataFrame이 된다.
- Pandas에서 사용하는 이러한 자료구조는 **리스트(list)**나 **딕셔너리(dict)**를 활용하여 직접 생성할 수 있다.

**리스트(list)**로 만들기

> `pd.Series()`로 Series형 자료를 만들 수 있다.

```python
mylist = [1,2,3,4]
mySeries = pd.Series(mylist)
mySeries
```

> `pd.DataFrame()`로 DataFrame형 자료를 만들 수 있다.

```python
mylists = [['James', 'Male', 'USA'],
           ['Bob', 'Male', 'UK'],
           ['Anna', 'Female', 'USA']]
df = pd.DataFrame(mylists)
df
# 컬럼명은 따로 지정해주지 않으면 0,1,2 순서의 인덱스로 설정된다.
```

> 컬럼명을 지정하고 싶은 경우 `columns` 옵션을 추가하면 된다.

```python
df = pd.DataFrame(mylists, columns=['이름','성별','나라'])
df
```

**딕셔너리(dict)**로 만들기
> 딕셔너리의 경우 리스트와 만드는 방법은 같으나,<br>key값을 바로 컬럼으로 인식하여 데이터프레임을 만들어주기 때문에 좀 더 편리하다.

```python
mydict = {'이름' : ['James', 'Bob', 'Anna'],
          '성별' : ['Male', 'Male', 'Female'],
          '나라' : ['USA', 'UK', 'USA']}
df = pd.DataFrame(mydict)
df
```

### 2-6-2. **loc**, **iloc** 을 사용해 편하게 조회하기
> - loc, iloc은 Pandas에서 데이터 조작을 한층 쉽게 만들어주는 매우 유용한 메소드이다.
- 축의 라벨(index)를 사용하여 데이터 읽기/쓰기, 슬라이싱 등 다양한 전처리를 할 수 있다.
- 반복문들 돌며 값을 검색하는 것이 아닌, index를 바로 찾아내는 방식(Vectorization)으로 매우 검색이 빠르다.
- 축의 이름으로 선택할 때는 **loc**, 정수 색인으로 선택할 때는 **iloc**을 사용한다.
- 사용 방법
    - `df.loc['행 슬라이싱 영역', '열 슬라이싱 영역']`
    - `df.iloc['행 슬라이싱 영역', '열 슬라이싱 영역']`

> #### loc
- 행과 열의 데이터를 조회할 때 특정 '레이블(label)'을 통해 선택하는 방법이다.
    - label은 인덱스와 거의 동일한 의미이다.
- location 의 약어로, 사람이 읽을 수 있는 label 값을 사용
- 직관적으로 label 이름만 적으면 데이터를 조회할 수 있다.

```python
# 현재 df의 행이 숫자index로 되어 있어 헷갈리기에 특정 label을 지닌 데이터로 바꾸겠다.
loc_df = df.copy()
loc_df.set_index('Korean Name', inplace=True)
loc_df
```

> **1. [loc] 특정 행 조회**

```python
# 특정 행 조회
loc_df.loc['임윤아']
```

```python
# 여러 행 조회
loc_df.loc[['임윤아', '김효연', '최수영']]
```

> **2. [loc] 조건문을 사용한 행 조회**

```python
# label 대신 조건문을 넣어 사용할 수도 있다.
df.loc[df['Korean Name']=='최수영']
```

```python
df.loc[df['Height']>=160]
```

> **3. [loc] 특정 열 조회**

```python
df.columns
```

```python
# 특정 열 조회
df.loc[:, ['Korean Name']]

# 하지만 단순 열 조회 같은 경우는 그냥 슬라이싱이 편해보인다. df['Korean Name']
```

```python
# 여러 열 조회
df.loc[:, ['Korean Name', 'Group']]
```

> **4. [loc] 조건문을 사용한 열 조회**

```python
# 열 조회 또한 마찬가지로 조건문을 이용할 수 있다.
df.loc[:, df.columns == 'Korean Name']
```

```python
df.loc[:, df.columns != 'Korean Name']
```

> **5. [loc] 행, 열 조건 조합**

```python
# 행과 열의 조건을 조합하여 사용할수도 있다.
df.loc[df['Korean Name']=='김효연', ['Height']]
```

```python
df.loc[df['Korean Name']=='김효연', df.columns == 'Height']
```

> #### iloc
- 행과 열의 데이터를 조회할 때 특정 '숫자 인덱스(integer index)'를 통해 선택하는 방법이다.
- interger location 의 약어로, 컴퓨터가 읽을 수 있는 index 값을 사용

> loc은 숫자든 문자든 해당 데이터의 label 값을 이용하여 데이터를 가져오는 것이고,<br> iloc은 label이 숫자든 문자든 상관없이 몇 번째냐를 기준으로 데이터를 가져오는 것

```python
iloc_df = loc_df.copy()
iloc_df
```

위의 데이터와 같은 상황에서 3번째의 값을 가져오고 싶어서

    iloc_df.loc[3]

### 2-6-3. Bollen Indexing, isin()

```python
df
```

> **1. Boolean Indexing**

> **2. isin() 을 활용한 색인**
- isin()은 데이터프레임에 알고 싶어서 지정한 자료가 들어있는지 여부를 확인해준다.
- 내가 정한 리스트(list) 안에 있는 값들만 가져오고 싶을 때 사용

```python
# 리스트를 지정
group_name = ['SNSD']
```

```python
# df의 'Group' 열에 group_name이 있는지 확인
df['Group'].isin(group_name)
```

```python
# isin()의 Boolean 값을 통해 해당하는 행 판별
df.loc[df['Group'].isin(group_name)]

# `df[df['Group'].isin(group_name)]`이라고 해도 된다.
```

    - 행 삭제 : `df.drop(index=['삭제하려는 행 index'])`

- 첫 번째 값은 행(row), 두 번째 값은 열(column)을 의미한다.

```python
df.shape
```

```python
df.head()
```

```python
df.head(3)
```

```python
df.tail()
```

```python
df.tail(2)
```

> `describe()` : 통계 정보 확인
- 주로 데이터 분석을 시작하기 전, 숫자형 데이터의 대략적 모습을 확인하려고 요약 통계 정보를 확인하기 위해 사용한다.
    - 통계적으로 계산을 할 수 있는 수치형 데이터에 대해서만 계산해준다.
- 보여주는 항목
    - 갯수(count), 평균(mean), 표준편차(std), 최소값(min), 1사분위수(25%), 2사분위수(50%), 3사분위수(75%), 최대값(max)

```python
df.describe()
```

### 2-7-2. 데이터프레임 정렬하기

> 인덱스(index) 기준 정렬
- `sort_index()` : 데이터프레임을 인덱스 기준으로 정렬한다.
    - 기본 default값은 오름차순이다.

```python
df.sort_index()
```

```python
# 내림차순 정렬 (`ascending=False` 옵션 추가)
df.sort_index(ascending=False)
```

> 값(value) 기준 정렬
- `sort_values()` : 데이터프레임을 값 기준으로 정렬한다.
    - 기본 default값은 오름차순이다.
    - 어떤 값을 기준으로 할지는 `by` 옵션을 추가하여 설정한다.

```python
df.sort_values(by='Height')
```

```python
# 특정 열 기준 내림차순 정렬
df.sort_values(by='Height', ascending=False)
```

> 복수 값(Multi-values) 기준 정렬
- `sort_values(by=['컬럼1','컬럼2', ...])` : 데이터프레임을 복수 값 기준으로 정렬한다.
- 첫 번째 값을 우선으로 정렬하여, 그 다음은 두 번째 값을 기준으로 차례로 정렬되는 형태이다.

```python
df.sort_values(by=['Height', 'Weight'])
```

```python
# 복수 값 기준 내림차순 정렬
df.sort_values(by=['Height', 'Weight'], ascending=False)
```

```python
# 복수 값의 기준을 각각 나누어서 내림차순 및 오름차순을 정할 수 있다.
df.sort_values(by=['Height','Weight'], ascending=[True,False])
```
