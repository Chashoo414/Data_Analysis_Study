## 셀 삽입
> - 바로 윗줄에 셀 생성 : Ctrl + M A (A: above)
> - 바로 아랫줄에 셀 생성 : Ctrl + M B (B: bottom)

## 셀 삭제
> - Ctrl + M D (D: delete)

## 셀 실행
> - 단일 셀 실행
  - 선택된 셀을 실행(커서는 그대로 남음) : Ctrl + Enter
  - 선택된 셀을 실행 후 다음 셀로 포커스 이동 : Shift + Enter
  - 선택된 셀 실행 후 다음줄에 새로운 셀 삽입 : Alt + Enter
> - 한번에 여러 셀 실행
  - Ctrl/Shift + 마우스 선택으로 실행하고자 하는 여러 셀 선택
    - Ctrl : 원하는 셀을 하나씩 선택하여 여러 셀을 선택
    - Shift : 현재 선택된 셀부터 마우스로 클릭한 셀까지 한번에 선택
  - 우클릭 후 '선택된 여러 셀 실행'
  - 단축키 : Ctrl + Shift + Enter
